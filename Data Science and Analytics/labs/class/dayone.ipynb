# creating continous data 

```python
#Method 1: from a python list 

temperatures = [36.6,37.2, 36.8,38.1, 36.5, 37.0, 36.9]
temp_series = pd.Series(temperatures, name = 'temperature')

print("== For a python list ==")
print(temp_series)
print(f"Data tyep : {temp_series.dtype}")
```

```python
# Method 2: from a numpy random generatoion 

np.random.seed(42)

heights = np.random.normal(loc=170, scale=10, size=100)

weights = np.random.normal(loc=65, scale=15, size=100)
salaries = np.random.normal(loc=50000, scale=10000, size=100)

print("== For numpy random generation ==")

print(f"Heights -means : {heights.mean():.2f}, std: {heights.std():.2f}")
print(f"Weights -means : {weights.mean():.2f}, std: {weights.std():.2f}")
print(f"Salaries -means : {salaries.mean():.2f}, Max: {salaries.std():.2f}")
```

```python
# Method 3 : Creating a DataFrame with multiple continous columns

df_contionus = pd.DataFrame({
    'Height_cm': np.round(heights,2),
    'Weight_kg': np.round(weights,2),
    'Salary': np.round(salaries,2),
    # 'temperature' : np.random.normal(36.8,0,5,100).round(1)
    
})


print("== DataFrame with multiple continuous columns ==")
# print(df_contionus.head(10))
# print(df_contionus.tail(10))

# print(f"\nData types : \n {df_contionus.dtypes}")
df_contionus
```

```python
# Creating Discrete data
```

```python
np.random.seed(42)

dice_rolls = np.random.randint(1,7,size=100)
goals_scored = np.random.poisson(lam=2, size=100)
coin_flips = np.random.binomial(n=1, p=0.5, size=100)


print("== Discrete data ==")
print(f"Dice rolls -uniwue values : {sorted(np.unique(dice_rolls))}")
print(f"Goals scored -range : goals_scored.min() to {goals_scored.max()}")
print(f"Coin flips -(heads in 10 ) -mean : coin_flips.mean()*10:.2f per 10 flips")
      
```

```python
#method 3 :  DataFrame with discrete columns 

df_discrete = pd.DataFrame({
    'dice_rolls': dice_rolls,
    'goals_scored': goals_scored,
    'heads_in_10_flips': coin_flips,
    'num_siblings' : np.random.choice([0,1,2,3,4], size=100, p=[0.2,0.4,0.3,0.1,0])
})

print("== DataFrame with discrete columns ==")
print(df_discrete.head(10))
print(f"\nData types : \n {df_discrete.dtypes}")
```

# Descriptive statistics 

```python
#  Individual statistics

print("== Individual statistics ==")
print(f"Mean height: {df_contionus['Height_cm'].mean():.2f} cm")
print(f"Median weight: {df_contionus['Weight_kg'].median():.2f} kg")
print(f"Standard deviation of salary: {df_contionus['Salary'].std():.2f}")
print(f"Mode of dice rolls: {df_discrete['dice_rolls'].mode()[0]}")
print(f"Variance of goals scored: {df_discrete['goals_scored'].var():.2f}")

print()

#percentiles and quantiles

q1 = df_contionus['Salary'].quantile(0.25)
q3 = df_contionus['Salary'].quantile(0.75)
iqr = q3 - q1
print(f"25th percentile (Q1) of salaries: {q1:.2f}")
print(f"75th percentile (Q3) of salaries: {q3:.2f}")
print(f"Interquartile Range (IQR) of salaries: {iqr:.2f}")


lower_bound = q1 - iqr*1.5
upper_bound = q3 + iqr*1.5
print(f"Lower bound for outliers: {lower_bound:.2f}")
print(f"Upper bound for outliers: {upper_bound:.2f}")
```
