```python
import matplotlib.pyplot as plt
regions = ['North', 'South', 'East', 'West']
revenue = [3000,2000,1000,2400]

plt.pie(revenue, labels  = regions, autopct='%1.1f%%', colors=['red','blue','green','yellow'])
plt.title('Revenue Contribution By reagion ')
plt.show()
```

# SSccatter Plot

### plt.scatter(x,y,color = 'color_name', marker = 'marker_symbol', label = 'label_name')



```python
hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [12,34,64,87,90,34,56,76]

plt.scatter(hours_studied, exam_scores, color = 'red', marker='d', label = 'Student scores')

plt.xlabel('hours_studied')
plt.ylabel('exam_scores')
plt.title('Distribution of marks')
plt.legend(loc = 'upper left')
plt.grid(True)

plt.show()
```

# subplot

### plt.subplot(nrow, ncol, index)



```python
x = [1,2,3,4,5,6,7]
y = [21,34,56,87,32,67,90]


plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('x vs y line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('X vx y bar chart')

# plt.tight_layout()
plt.show()

```

```python
#fig, ax = plt.subplots(nrows, ncols, figsize = (width, height))

fig,ax = plt.subplots(1,2, figsize =(10,5))

x = [1,2,3,4,5,6,7]
y = [21,34,56,87,32,67,90]


ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Graph')

fig.suptitle('Comparison of line and Bar Charts')

plt.tight_layout()
plt.show()
```

# Saving the Figures 
### plt.savefig('Filename.ext', dpi = value, bbox _inches = 'tight)

```python

fig,ax = plt.subplots(1,2, figsize =(10,5))

x = [1,2,3,4,5,6,7]
y = [21,34,56,87,32,67,90]


ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Graph')

fig.suptitle('Comparison of line and Bar Charts')

plt.tight_layout()

plt.savefig('lineplot.png', dpi = 500, bbox_inches = 'tight')
plt.show()
```
