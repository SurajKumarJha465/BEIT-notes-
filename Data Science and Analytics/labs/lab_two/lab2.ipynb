### 1. Creating 1-D and 2-D Arrays Using NumPy

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
print("1-D Array:")
print(arr1)

arr2 = arr1.reshape(1, 5)
print("Reshaped 2-D Array:")
print(arr2)

arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[5, 6], [7, 8]])

sum_array = arr3 + arr4

print("Addition of Arrays:")
print(sum_array)

print("Shape of arr3:")
print(arr3.shape)
```

### 2. Accessing the Last Element of a 2-D Array

```python
print(arr3[-1, -1])
```

### 3. Array Slicing

```python
print(arr1[1:4])

print(arr3[:, 1])

print(arr1[::-1])
```

### 4. Searching Data Using where() and searchsorted()

```python
data = np.array([10, 20, 30, 40, 50])

x = np.where(data == 30)
print(x)

y = np.searchsorted(data, 35)
print(y)
```

### 5. Statistical Calculations

```python
from statistics import mode
import numpy as np

data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
median = np.median(data)
mode_value = mode(data)
std = np.std(data)
variance = np.var(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode_value)
print("Standard Deviation:", std)
print("Variance:", variance)

actual = np.array([10, 20, 30])
predicted = np.array([12, 18, 33])

mse = np.mean((actual - predicted) ** 2)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```

### 6. Creating DataFrame Using Dictionaries and NumPy Arrays

### 9. Data Visualization Using Matplotlib

```python
import matplotlib.pyplot as plt

subjects = [
    "Probability and Statistics",
    "Data Science and ML",
    "Artificial Intelligence",
    "Computer Graphics",
    "Engineering Economics",
    "Research Fundamentals"
]

marks = [60, 45, 59, 72, 80, 91]

# Line Plot
plt.plot(subjects, marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Line Plot")
plt.xticks(rotation=15)
plt.show()

# Scatter Plot
plt.scatter(subjects, marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Scatter Plot")
plt.xticks(rotation=15)
plt.show()

# Bar Graph
plt.bar(subjects, marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Bar Graph")
plt.xticks(rotation=15)
plt.show()

# Pie Chart
plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
